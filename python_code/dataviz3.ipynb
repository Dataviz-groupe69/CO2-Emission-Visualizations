```python
from sklearn import linear_model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./annual_csv.csv",sep=",")
data_to_numpy = data.values
```

```python
data.head
```

```python
data_to_numpy
```

```python
x = data_to_numpy[:,0]
y_GCAG = data_to_numpy[:,1]
```

```python
regr = linear_model.Lasso()
regr.fit(x[75:,np.newaxis], y_GCAG[75:])
```

```python
x_test = np.arange(2017,2051)
```

```python
predict = regr.predict(x_test[:,np.newaxis])
```

```python
predict
```

```python
y_GISTEMP = data_to_numpy[:,2]
```

```python
regr_GISTEMP = linear_model.Lasso()
regr_GISTEMP.fit(x[75:,np.newaxis], y_GISTEMP[75:])
```

```python
predict_GISTEMP = regr_GISTEMP.predict(x_test[:,np.newaxis])
```

```python
predict_GISTEMP
```

```python
for i in range(len(x_test)):
    data_to_numpy = np.vstack([data_to_numpy,[x_test[i],predict[i],predict_GISTEMP[i]]])
```

```python
data_to_numpy
```

```python
data_to_numpy=data_to_numpy[data_to_numpy[:,0]>=1950]
```
