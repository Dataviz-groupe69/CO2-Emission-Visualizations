```python
def algorithme_remplissage(data_base):
    pays = set(data_base["Country"])
    numpy_data_pays = data_base.values
    data_tmp = numpy_data_pays[:,1]
    pay_int = pay_en_int(pays)
    for pay_ in pays:
        data_tmp[data_tmp==pay_]=pay_int[pay_] 
    numpy_data_pays[:,1] = data_tmp
    for pay in pays :
        regr = linear_model.Lasso()
        numpy_data = numpy_data_pays[numpy_data_pays[:,1]==pay_int[pay]]
        #numpy_data = numpy_data_1[numpy_data_1[:,0]>1950]
        x_test = []
        x = numpy_data[:,0]
        y = numpy_data[:,2]
        for i in np.arange(1950,2051):
            if( i not in x):
                x_test.append(i)
        x_test_1 = np.array(x_test)
        print(len(x),len(y))
        regr.fit(x[:,np.newaxis], y)
        predict = regr.predict(x_test_1[:,np.newaxis])
        predict.astype(int)
        for i in range(len(predict)):
            numpy_data_pays = np.vstack([numpy_data_pays,[x_test_1[i],pay_int[pay],predict[i]]])
    return numpy_data_pays
```

```python
def ecrire_json(resultat):
    import json
    json = json.dumps(resultat)
    f = open("./final.json","w")
    f.write(json)
    f.close()
```

```python
def pay_en_int(pays):   
    pays_to_int = {}

    for pay_iter,i in zip(pays,range(len(pays))):
        pays_to_int[pay_iter] = i
    return pays_to_int
```
